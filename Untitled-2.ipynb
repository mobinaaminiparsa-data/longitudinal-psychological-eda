```python
num_cols = ['day','DailyStressor','EFcoping','NeA','PSS10','TraitAnxiety','RES']

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')


df[num_cols] = df[num_cols].replace(999, np.nan)

df[num_cols] = df.groupby('ID')[num_cols].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df['ID'] = df['ID'].astype(str)
df = df.sort_values(['ID','day']).reset_index(drop=True)

print(df.head())
```

# 9

```python
person_features = df.groupby("ID").agg({
    "DailyStressor":["mean","std"],
    "NeA":["mean","std"],
    "PSS10":"mean",
    "RES":"mean"
})

person_features.columns = ["_".join(col) for col in person_features.columns]
```

# 10

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(person_features)
```

```python
person_features.isnull().sum()
person_features[person_features.isnull().any(axis=1)]
person_features = person_features.dropna()
```

# 11

```python
from sklearn.cluster import KMeans

inertia = []

for k in range(1,10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_data)
    inertia.append(km.inertia_)

plt.plot(range(1,10), inertia, marker="o")
plt.show()
```

# 12

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

person_features["Cluster"] = clusters
```

# 13

```python
df = df.merge(person_features["Cluster"], on="ID")
```

## Radar plot

```python
cluster_profile = person_features.groupby("Cluster").mean()

categories = cluster_profile.columns
N = len(categories)

angles = np.linspace(0, 2*np.pi, N, endpoint=False)
angles = np.concatenate([angles, [angles[0]]])

plt.figure(figsize=(8,8))

for i in cluster_profile.index:
    values = cluster_profile.loc[i].values
    values = np.concatenate([values, [values[0]]])
    
    plt.polar(angles, values, label=f"Cluster {i}")

plt.xticks(angles[:-1], categories)
plt.title("Cluster Behavioral Profiles")
plt.legend()
plt.show()
```
